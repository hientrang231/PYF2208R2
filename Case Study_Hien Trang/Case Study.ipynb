Quản lý danh sách học viên tại Codegym

```python
class Student:
    def __init__(self, name, age, sex, course, email):
        self.name=name
        self.age=age
        self.sex=sex
        self.course=course
        self.email=email
```

```python
from tkinter import *
from tkinter import messagebox
from tkinter import ttk
from PIL import Image,ImageTk

window = Tk()
img  = Image.open('/Users/hientrang/Desktop/background.png') 
photo=ImageTk.PhotoImage(img)
lab=Label(image=photo).place(x=0, y=0)
window.geometry('1000x300')


window.title('Student record management at Codegym system')
list_student = []
window['bg'] = '#bbdef9'
bg = PhotoImage(file = '/Users/hientrang/Desktop/background.png')
 
    
    
# Show image using label
name_lbl = Label(window, text = 'Name: ', bg ='#63B8FF', fg='black', bd=0)
age_lbl = Label(window, text = 'Age: ', bg ='#63B8FF', fg='black', bd=0)
sex_lbl = Label(window, text = 'Sex: ', bg ='#63B8FF', fg='black', bd=0)
email_lbl = Label(window, text = 'Email: ', bg ='#bbdef9', fg='black', bd=0)
course_lbl = Label(window, text = 'Course: ', bg ='#bbdef9', fg='black', bd=0)


name = Entry(window, bg = '#bbdef9', fg='black')
age = Entry(window, bg ='#bbdef9', fg='black')
sex =  Entry(window, bg ='#bbdef9', fg='black')
email = Entry(window, bg ='#bbdef9', fg='black')
course = Entry(window, bg ='#bbdef9', fg='black')

check = Entry(window, bg ='#bbdef9', fg='black')

#add
def add_student():
    list_student.append(Student(name.get(), int(age.get()), sex.get(), email.get(), course.get()))
    print([item.name for item in list_student])
    messagebox.showinfo('Student', f'{name.get()} added sucessfully')
    refresh_data(list_student)
    
#search  
def search_student():
    search_name=check.get()
    exist=False 
    for item in list_student:
        if item.name==search_name:  
            exist=True 
            messagebox.showinfo('Student', f'{item.name} is in list with: Age is {item.age}, Sex is {item.sex}, Email is {item.email}, Course is {item.course}')
            break
        if exist==False:
            messagebox.showinfo('Student', f'{check.get()} does not exist in list student')    
    refresh_data(list_student)
    
#delete
def delete_student():
    delete_name=check.get()
    exist=False
    for i in range(len(list_student)):
        if list_student[i].name==delete_name:
            exist=True
            del list_student[i]
            messagebox.showinfo('Student', f'{delete_name} is deleted sucessfully')                
            break
        if exist==False:
            messagebox.showinfo('Student', f'{check.get()} does not exist in list student')

    refresh_data(list_student)
#update
def update_student():
    update_name = check.get()
    search_name = check.get()
    exist = False 
    for item in list_student:
        if item.name == search_name: 
            exist = True
            item.name = name.get()
            item.age = age.get()
            item.sex = sex.get()
            item.course = course.get()
            item.email = email.get()
            messagebox.showinfo('Student',f'{name.get()} has been updated succesfully')
            break
    if exist == False:
        messagebox.showinfo('Student',f'{check.get()} does not exist in list')   
    print([item.name for item in list_student])

    refresh_data(list_student)

add_btn = Button(window, text = 'Add', command=add_student, fg='blue', borderwidth=0)
delete_btn = Button(window, text = 'Delete', command=delete_student, fg='blue', borderwidth=0)
search_btn = Button(window, text = 'Search', command=search_student, fg='blue', borderwidth=0)
update_btn = Button(window, text = 'Update', command=update_student, fg='blue', borderwidth=0)

    
name_lbl.place(x = 20, y = 15)
age_lbl.place(x = 20, y = 45)
sex_lbl.place(x = 20, y = 75)
email_lbl.place(x = 20, y = 105)
course_lbl.place(x = 20, y = 135)

name.place(x = 110, y = 10)
age.place(x = 110, y = 40)
sex.place(x= 110, y = 70)
email.place(x=110, y = 100)
course.place(x = 110, y= 130)

add_btn.place(x = 170, y = 167)
update_btn.place(x = 110, y = 250)
search_btn.place(x = 20, y = 200)
delete_btn.place(x = 228, y= 250)
check.place(x = 110, y = 200)

#tree
def refresh_data(list_student): 
    print(len(list_student))
    for row in tree.get_children():
        tree.delete(row)
    for i in range(0,len(list_student)):
        tree.insert('', END, values=(
            list_student[i].name,
            list_student[i].email,
            list_student[i].course
        ))
            
def item_selected(event):
    pass
            
columns = ('name', 'email', 'course')
tree = ttk.Treeview(columns = columns, show = 'headings')
tree.heading('name', text='Name')
tree.heading('email', text='Email')
tree.heading('course', text='Course')

tree.bind('<<TreeviewSelect>>', item_selected)
tree.place(x = 350, y = 10)
refresh_data(list_student)

window.mainloop()
```
