# Bai 3
Nhap vao 'tien_luong_theo_gio' va 'so_gio' trong 1 thang, va 'tien_thuong'. Sau do in ra tong thu nhap trong thang.
Biet rang: Tong thu nhap = tien luong theo gio + tien thuong

```python
tien_luong_theo_gio=float(input("Nhap tien luong theo gio: "))
so_gio=float(input("Nhap so gio: "))
tien_thuong=float(input("Nhap so tien thuong: "))
tong_thu_nhap=tien_luong_theo_gio*so_gio+tien_thuong
print(f"Tong thu nhap trong thang la {tong_thu_nhap}")
```

# Bai 4
Nhap vao mot String va mot so x. Hay xoa x ky tu dau tien cua string
Vdu: string = 'Hello Python', x=3
    Output: 'lo Python'

```python
string=input('Nhap string: ')
x=int(input('Nhap so ky tu can xoa: '))
string=string[x:]
print(f'String sau khi cat la: {string}')
```
